```
import httpx
import pandas as pd

# Initialize an empty dataframe to store the cinema data
cinema_df = pd.DataFrame(columns=['theatreId', 'name', 'latitude', 'longitude', 'oldCinemaCode'])

def fetch_cinemas_data():
    # Target URL
    url = "https://api3.pvrcinemas.com/api/v1/booking/content/cinemas"

    # Request headers
    headers = {
        "Host": "api3.pvrcinemas.com",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:133.0) Gecko/20100101 Firefox/133.0",
        "Accept": "application/json, text/plain, */*",
        "Accept-Language": "en-US,en;q=0.5",
        "Accept-Encoding": "gzip, deflate, br, zstd",
        "Content-Type": "application/json",
        "Authorization": "Bearer ",  # Add the Bearer token if required
        "chain": "PVR",
        "city": "Bengaluru",
        "appVersion": "1.0",
        "platform": "WEBSITE",
        "country": "INDIA",
        "Origin": "https://www.pvrcinemas.com",
        "DNT": "1",
        "Connection": "keep-alive",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-site",
        "Priority": "u=0"
    }

    # Request payload
    payload = {
        "city": "Bengaluru",
        "lat": "",
        "lng": "",
        "text": ""
    }

    try:
        # Use an HTTP/2 client to mimic browser behavior
        with httpx.Client(http2=True) as client:
            response = client.post(url, headers=headers, json=payload)

            # Check the response status
            if response.status_code == 200:
                print("Response Data Retrieved Successfully!")
                response_data = response.json()

                # Extract cinemas data
                cinemas = response_data.get('output', {}).get('c', [])

                # Filter the relevant information: theatreId, name, latitude, longitude, oldCinemaCode
                filtered_cinemas = []
                for cinema in cinemas:
                    filtered_cinema = {
                        'theatreId': cinema.get('theatreId'),
                        'name': cinema.get('name'),
                        'latitude': cinema.get('latitude'),
                        'longitude': cinema.get('longitude'),
                        'oldCinemaCode': cinema.get('oldCinemaCode')
                    }
                    filtered_cinemas.append(filtered_cinema)

                # Convert filtered data to pandas DataFrame
                global cinema_df
                cinema_df = pd.DataFrame(filtered_cinemas)

                # Print the DataFrame
                print(cinema_df)
            else:
                print(f"Request failed with status code: {response.status_code}")
                print("Response Body:")
                print(response.text)
    except Exception as e:
        print(f"An error occurred: {e}")

# Run the function to fetch cinema data
fetch_cinemas_data()

# Example: Access the DataFrame
print(cinema_df)
```
